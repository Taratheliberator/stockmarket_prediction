```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python

data = pd.read_csv('Прошлые данные - SBER.csv', delimiter=',', quotechar='"')
```

```python
print(data.head())
```

```python

data['Дата'] = pd.to_datetime(data['Дата'], format='%d.%m.%Y')
data['Цена'] = data['Цена'].str.replace(',', '.').astype(float)


data_sorted = data.sort_values(by='Дата')


plt.figure(figsize=(15, 7))
plt.plot(data_sorted['Дата'], data_sorted['Цена'], label='Closing Price', color='blue')
plt.title('Closing Price of SBER over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

```python

def convert_volume_to_float(volume):
    if 'M' in volume:
        return float(volume.replace('M', '').replace(',', '.')) * 1e6
    else:
        return float(volume.replace(',', '.'))
```

```python

if data_sorted['Объём'].dtype == 'object':
    data_sorted['Объём'] = data_sorted['Объём'].apply(convert_volume_to_float)
data_sorted['Объём']   
```

```python

for column in ['Откр.', 'Макс.', 'Мин.']:
    if data_sorted[column].dtype == 'object':
        data_sorted[column] = data_sorted[column].str.replace(',', '.').astype(float)


if data_sorted['Изм. %'].dtype == 'object':
    data_sorted['Изм. %'] = data_sorted['Изм. %'].str.replace(',', '.').str.rstrip('%').astype(float) / 100.0
data_sorted
```

```python
# Create lagged features
for i in range(1, 6):
    data_sorted[f'lag_{i}'] = data_sorted['Цена'].shift(i)

# Drop NaN values (which were introduced because of the lagged features)
data_sorted = data_sorted.dropna()

# Split data into training and test sets (80% train, 20% test)
train_size = int(0.8 * len(data_sorted))
train, test = data_sorted[:train_size], data_sorted[train_size:]


X_train = train.drop(columns=['Дата', 'Цена'])
y_train = train['Цена']
X_test = test.drop(columns=['Дата', 'Цена'])
y_test = test['Цена']


rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)


y_pred = rf_model.predict(X_test)


mse = mean_squared_error(y_test, y_pred)

mse
```

```python

plt.figure(figsize=(15, 7))
plt.plot(test['Дата'], y_test, label='Actual', color='blue')
plt.plot(test['Дата'], y_pred, label='Predicted', color='red', linestyle='--')
plt.title('Actual vs Predicted Closing Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

```python

feature_importances = rf_model.feature_importances_


features_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': feature_importances
})


features_df = features_df.sort_values(by='Importance', ascending=False)


plt.figure(figsize=(12, 8))
plt.barh(features_df['Feature'], features_df['Importance'], color='skyblue')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importances from Random Forest Model')
plt.gca().invert_yaxis()  # Highest importance at the top
plt.show()
```

```python
# Function to generate predictions for the next days
def predict_next_days(model, data, days=5):
    predictions = []
    input_data = data[-5:].copy()  
    for _ in range(days):
        
        features = input_data.iloc[-1][['lag_1', 'lag_2', 'lag_3', 'lag_4', 'lag_5', 
                                       'Откр.', 'Макс.', 'Мин.', 'Объём', 'Изм. %']].values.reshape(1, -1)
       
        predicted_value = model.predict(features)[0]
       
        predictions.append(predicted_value)
        
        new_row = input_data.iloc[-1].copy()
        new_row['lag_5'] = new_row['lag_4']
        new_row['lag_4'] = new_row['lag_3']
        new_row['lag_3'] = new_row['lag_2']
        new_row['lag_2'] = new_row['lag_1']
        new_row['lag_1'] = predicted_value
        new_row['Дата'] = new_row['Дата'] + pd.Timedelta(days=1)
    
        input_data = input_data.append(new_row, ignore_index=True)
    return predictions


next_5_days_predictions = predict_next_days(rf_model, data_sorted, days=5)
```

```python


plt.figure(figsize=(15, 7))


plt.plot(data_sorted['Дата'][-30:], data_sorted['Цена'][-30:], label='Actual', color='blue')


next_5_dates = [data_sorted['Дата'].max() + pd.Timedelta(days=i) for i in range(1, 6)]
plt.plot(next_5_dates, next_5_days_predictions, label='Predicted', color='red', linestyle='--', marker='o')

plt.title('Actual vs Predicted Closing Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor


dt_model = DecisionTreeRegressor(random_state=42)


dt_model.fit(X_train, y_train)


y_pred_dt = dt_model.predict(X_test)


mse_dt = mean_squared_error(y_test, y_pred_dt)

mse_dt
```

```python
from sklearn.ensemble import GradientBoostingRegressor


gbt_model = GradientBoostingRegressor(n_estimators=100, random_state=42)


gbt_model.fit(X_train, y_train)


y_pred_gbt = gbt_model.predict(X_test)


mse_gbt = mean_squared_error(y_test, y_pred_gbt)

mse_gbt
```

```python

plt.figure(figsize=(15, 7))
plt.plot(test['Дата'], y_test, label='Actual', color='blue')
plt.plot(test['Дата'], y_pred_gbt, label='Predicted', color='red', linestyle='--')
plt.title('Actual vs Predicted Closing Prices (Gradient Boosted Trees)')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```
